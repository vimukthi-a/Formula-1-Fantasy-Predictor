```python
import pandas as pd
```

```python
v2 = pd.read_csv("../results/fantasy_v2_results.csv")
v3 = pd.read_csv("../results/fantasy_v3_results.csv")

# print(v2.columns)
# print(v3.columns)
```

```python
v2[["RaceName", "Average Actual Finish", "Average Prediction Error"]].head()
```

```python
v3[["RaceName", "Average Actual Finish", "Average Prediction Error"]].head()
```

```python
comparison = pd.DataFrame({
    "RaceName": v2["RaceName"],

    "V2 Average Finish": v2["Average Actual Finish"],
    "V3 Average Finish": v3["Average Actual Finish"],

    "V2 Top5": v2["Top 5 hit rate"],
    "V3 Top5": v3["Top 5 hit rate"],

    "V2 Top10": v2["Top 10 hit rate"],
    "V3 Top10": v3["Top 10 hit rate"],

    "V2 Picks": v2["Picks"],
    "V3 Picks": v3["Picks"]  
})

comparison
```

```python
comparison["Finish difference"] = (
    comparison["V3 Average Finish"] - 
    comparison["V2 Average Finish"]
)

comparison["Top 5 Difference"] = (
    comparison["V3 Top5"] - 
    comparison["V2 Top5"]
)

comparison["Top 10 Difference"] = (
    comparison["V3 Top10"] - 
    comparison["V2 Top10"]
)

comparison["Changed Picks"] = (
    comparison["V3 Picks"] != comparison["V2 Picks"]
)

comparison

# Results not bad, personally I do not like the decimals so im gonna change it up
```

```python
# print(comparison.round(3).head())
```

```python
def driver_pick_changes(v2_picks, v3_picks):
    v2_list = [pick.strip() for pick in v2_picks.split(",")]
    v3_list = [pick.strip() for pick in v3_picks.split(",")]


    v2_set = set(v2_list)
    v3_set = set(v3_list)

    removed = sorted(v2_set - v3_set)
    added = sorted(v3_set - v2_set)

    if removed or added:
        return f"Removed {', '.join(removed)} | Added: {', '.join(added)}"
    
    if v2_list != v3_list:
        return "Same drivers, different order"
    
    return "False"

# This function is by far the most useful one here, so we know the results will fluctuate a bit and because it is hard to read side by side
# This function will basically use the set function to check if the values picked are same or not
# Set makes our life easy #set(s)4life
# Some rows have same drivers but different order while others have no change
# But the ones with changes we will say who we removed and who we replaced them with
```

```python
comparison["Pick Changes"] = comparison.apply(
    lambda row: driver_pick_changes(row["V2 Picks"], row["V3 Picks"]),
    axis = 1
)

comparison = comparison.round(3)

comparison
```

Improved the overall readability of the table and the changed picks and everything

False is the same as same driver, different order because we have no change and its just simply the same thing. 
Since we can see only 3 out of 22 being changed, this means that the overall foundation of this model is not destroyed nor is GridGap a dominating feature which changes the whole model entirely.
